```python
import pandas as pd
data = '../leuven_bce_transposed.csv'

df = pd.read_csv(data, index_col=0)
#calculate the number of rows
# total number of entries in each row that are 0
df['zeros'] = df.apply(lambda x: x.eq(0).sum(), axis=1)

# total number of entries in each row that are 1
df['ones'] = df.apply(lambda x: x.eq(1).sum(), axis=1)

df['total'] = df['zeros'] + df['ones']
# remove rows that have 86 zeros or 86 ones
df = df[(df['zeros'] != 86) & (df['ones'] != 86)]

# calculate positive weight for binary cross entropy
df['pos_weight'] = df['zeros'] / df['ones']
df
```

```python
df.to_csv('../leuven_bce_transposed_clean_with_pos_weight.csv')
```
